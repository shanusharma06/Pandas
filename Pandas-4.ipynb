## Binning

```python
weather_tidy
```

```python
weather_tidy['max'].min()
```

```python
weather_tidy['max'].max()
```

```python
bins = [16,20,24]
labels = ['low','high']

weather_tidy['max_category'] = pd.cut(weather_tidy['max'],
                                     bins = bins,
                                     labels = labels)
weather_tidy.head(10)
```

```python
bins = [19,20,24]
labels = ['low','high']

weather_tidy['max_category2'] = pd.cut(weather_tidy['max'],
                                     bins = bins,
                                     labels = labels)
weather_tidy.head(10)
```
